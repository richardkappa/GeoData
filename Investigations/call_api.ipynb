```python
import requests
import pandas as pd
import os
import sys


import geopandas as gpd

%matplotlib inline
import matplotlib.pyplot as plt
plt.ioff()



parentdir = os.path.dirname(os.getcwd())
sys.path.insert(0, parentdir)
import KeyFunctions as kf

root_path = "D:/GeoData/"
Main_CRS = "EPSG:27700"
```

```python
#Get UK map
gdf = kf.load_obj(root_path,"raw_gdf")
```

```python
api_url = "http://environment.data.gov.uk/flood-monitoring/id/floods?min-severity=2"
response = requests.get(api_url)
flood_warnings = pd.json_normalize(response.json()['items'])
```

```python
boundary = gpd.read_file(flood_warnings["floodArea.polygon"][0])
boundary = boundary.to_crs(Main_CRS)
boundary
```

```python
boundaries = pd.DataFrame(columns=gpd.read_file(flood_warnings["floodArea.polygon"][0]).columns)
for i in flood_warnings["floodArea.polygon"]:
    bdry = gpd.read_file(i)
    boundaries = pd.concat([boundaries, bdry], ignore_index=True)
    boundaries_gdf = gpd.GeoDataFrame(boundaries, crs=Main_CRS, geometry="geometry")
```
